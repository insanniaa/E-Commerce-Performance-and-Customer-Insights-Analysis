# **Menentukan Pertanyaan Bisnis**

Kita perlu memahami permasalahan bisnis yang dihadapi terlebih dahulu. Pada proyek ini, kita akan menganalisis data dari sebuah perusahaan e-commerce yang menjual berbagai produk secara online. Perusahaan ini membutuhkan insight untuk meningkatkan performa bisnisnya, termasuk penjualan, kepuasan pelanggan, dan efisiensi pengiriman.

Sebagai perusahaan yang terus berkembang di dunia digital, e-commerce ini perlu mengevaluasi berbagai aspek seperti performa penjualan produk, pola belanja pelanggan, dan efisiensi pengiriman produk. Selain itu, perusahaan juga perlu memahami ulasan pelanggan untuk meningkatkan kualitas layanan dan produk.

Berdasarkan beberapa permasalahan tersebut, berikut adalah beberapa pertanyaan bisnis yang dapat dijawab melalui analisis data:

1. Bagaimana tren penjualan (sales) dan pendapatan (revenue) perusahaan dalam beberapa bulan terakhir?
   - Analisis ini akan membantu perusahaan memahami fluktuasi penjualan, mengidentifikasi bulan-bulan dengan performa penjualan tertinggi dan terendah, serta mengevaluasi faktor-faktor yang mempengaruhi penurunan atau peningkatan pendapatan.

2. Produk apa yang paling banyak dan paling sedikit terjual?
   - Dengan mengetahui produk yang paling laris, perusahaan dapat fokus pada pemasaran produk-produk populer. Sebaliknya, produk yang jarang terjual dapat dievaluasi lebih lanjut untuk pengurangan stok atau perbaikan strategi pemasaran.

3. Bagaimana distribusi demografi pelanggan yang kita miliki?
   - Penting untuk mengetahui karakteristik pelanggan, seperti lokasi geografis, usia, atau kategori lainnya, guna menargetkan kampanye promosi secara lebih tepat dan efektif.

4. Kapan terakhir kali pelanggan melakukan transaksi?
   - Dengan memahami waktu transaksi terakhir dari setiap pelanggan, perusahaan dapat mengidentifikasi pelanggan yang belum melakukan pembelian dalam jangka waktu tertentu dan menjalankan strategi re-engagement.

5. Seberapa sering seorang pelanggan melakukan pembelian dalam beberapa bulan terakhir?
   - Frekuensi pembelian membantu perusahaan memahami loyalitas pelanggan dan mengidentifikasi pelanggan tetap serta pelanggan yang cenderung hanya melakukan pembelian sekali.

6. Berapa banyak uang yang dihabiskan pelanggan dalam beberapa bulan terakhir?
   - Dengan analisis ini, perusahaan bisa mengetahui pola pengeluaran pelanggan, mengidentifikasi pelanggan dengan nilai tinggi (high-value customers), dan menargetkan mereka dengan kampanye eksklusif atau promosi khusus.

7. Produk apa yang memiliki rating paling tinggi dan rendah?
  - Analisis ini bertujuan untuk memahami bagaimana pelanggan menilai produk-produk yang telah mereka beli. Produk dengan rating tinggi menunjukkan kepuasan pelanggan yang baik dan mungkin bisa dijadikan andalan dalam pemasaran. Sebaliknya, produk dengan rating rendah perlu dievaluasi lebih lanjut untuk mengidentifikasi masalah yang mungkin terjadi, seperti kualitas, deskripsi produk yang kurang akurat, atau pelayanan pengiriman yang kurang memuaskan. Dengan demikian, perusahaan bisa mengambil tindakan korektif, seperti meningkatkan kualitas produk atau memberikan penjelasan yang lebih detail mengenai produk untuk menyesuaikan ekspektasi pelanggan.

### Menghitung waktu pengiriman tiap order

```python
# Menghitung waktu pengiriman (delivery time) dengan mengurangi tanggal pengiriman ke pelanggan
# dengan tanggal pengiriman ke pengangkut. Hasilnya adalah objek timedelta.
delivery_time = orders_df["order_delivered_customer_date"] - orders_df["order_delivered_carrier_date"]

# Mengonversi timedelta ke detik total dan menyimpannya dalam variabel delivery_time.
# Dengan menggunakan fungsi apply, kita menghitung total detik untuk setiap entri.
delivery_time = delivery_time.apply(lambda x: x.total_seconds())

# Menghitung waktu pengiriman dalam hari dengan membagi total detik dengan jumlah detik dalam satu hari (86400 detik).
# Hasilnya dibulatkan ke bilangan bulat terdekat dan disimpan dalam kolom baru "delivery_time" pada DataFrame orders_df.
orders_df["delivery_time"] = round(delivery_time / 86400)
```

### Mengetahui produk yang paling sering dibeli

```python
# Menggabungkan (merge) dua DataFrame: order_items_df dan products_df.
# DataFrame order_items_df berisi informasi item pesanan, sedangkan products_df berisi informasi produk.
sales_product_df = pd.merge(
    left=order_items_df,        # DataFrame di sebelah kiri (left) yang berisi data item pesanan
    right=products_df,          # DataFrame di sebelah kanan (right) yang berisi data produk
    how="left",                 # Metode penggabungan yang digunakan adalah left join, yang akan mengembalikan semua baris dari order_items_df dan mencocokkan baris yang relevan dari products_df.
    left_on="product_id",       # Kolom pada DataFrame sebelah kiri yang digunakan untuk mencocokkan (join)
    right_on="product_id"       # Kolom pada DataFrame sebelah kanan yang digunakan untuk mencocokkan (join)
)

# Menampilkan lima baris pertama dari DataFrame hasil penggabungan sales_product_df.
sales_product_df.head()
```

```python
# Mengelompokkan DataFrame sales_product_df berdasarkan 'product_id' dan 'product_category_name'.
# Pengelompokan ini akan memungkinkan kita untuk menghitung jumlah pembelian dan total pendapatan
# untuk setiap produk dalam kategori tertentu.
product_purchase_count = sales_product_df.groupby(['product_id', 'product_category_name']).agg(
    purchase_count=('order_id', 'size'),    # Menghitung jumlah pesanan (order_id) untuk setiap kelompok, yang diwakili sebagai 'purchase_count'.
    total_revenue=('price', 'sum')          # Menghitung total pendapatan (sum dari harga 'price') untuk setiap kelompok, yang diwakili sebagai 'total_revenue'.
).reset_index()  # Mengatur ulang indeks agar hasil pengelompokan menjadi DataFrame biasa.

# Mengurutkan DataFrame product_purchase_count berdasarkan kolom 'purchase_count' dalam urutan menurun agar produk dengan jumlah pembelian tertinggi berada di atas.
product_purchase_count.sort_values(by='purchase_count', ascending=False)
```

Dapat diketahui bahwa produk yang paling sering dibeli adalah produk id aca2eb7d00ea1a7b8ebd4e68314663af dari kategori moveis decoracao dengan jumlah pembelian 527 kali

### Melihat informasi penjualan berdasarkan kategori produknya

```python
category_purchase_summary = product_purchase_count.groupby(by="product_category_name")[['purchase_count', 'total_revenue']].sum().reset_index().sort_values(by='purchase_count', ascending=False)
print(category_purchase_summary)
```

Dapat diketahui bahwa kategori produk yang paling laris adalah cama mesa banho dengan pembelian sebanyak 11115 produk.

```python
category_purchase_summary = product_purchase_count.groupby(by="product_category_name")[['purchase_count', 'total_revenue']].sum().reset_index().sort_values(by='total_revenue', ascending=False)
print(category_purchase_summary)
```

Dapat diketahui bahwa kategori produk yang menyumbang revenue toko paling banyak adalah baleza_saude yaitu dengan revenue sebesar 1258681,34 USD.

## Eksplorasi data order_reviews_df

### Menghitung seberapa cepat ulasan ditanggapi

```python
# Menghitung waktu respons (response time) dengan mengurangi waktu penjawaban ulasan (review_answer_timestamp)
# dengan waktu pembuatan ulasan (review_creation_date). Hasilnya adalah objek timedelta.
response_time = order_reviews_df["review_answer_timestamp"] - order_reviews_df["review_creation_date"]

# Mengonversi timedelta ke total detik dan menyimpannya dalam variabel response_time.
# Dengan menggunakan fungsi apply, kita menghitung total detik untuk setiap entri.
response_time = response_time.apply(lambda x: x.total_seconds())

# Menghitung waktu respons dalam hari dengan membagi total detik dengan jumlah detik dalam satu hari (86400 detik).
# Hasilnya dibulatkan ke bilangan bulat terdekat dan disimpan dalam kolom baru "response_time" pada DataFrame order_reviews_df.
order_reviews_df["response_time"] = round(response_time / 86400)
```

## Menggabungkan semua tabel

```python
# Menggabungkan orders_df dengan customers_df
merged_orders_sales = pd.merge(
    left=orders_df,
    right=customers_df,
    how='inner',
    on='customer_id'
)

# Menggabungkan order_items_df dengan sellers_df
merge_order_items_sellers = pd.merge(
    left=order_items_df,
    right=sellers_df,
    how='inner',
    on='seller_id'
)

# Menggabungkan order_items_df dengan products_df
merge_order_items_sellers_products = pd.merge(
    left=merge_order_items_sellers,
    right=products_df,
    how='inner',
    on='product_id'
)

# Menggabungkan hasil merge_orders_sales dengan merge_order_items_sellers_products
merged_orders_sales_customers_sellers_products = pd.merge(
    left=merged_orders_sales,
    right=merge_order_items_sellers_products,
    how='inner',
    on='order_id'
)

# Menggabungkan hasil di atas dengan installment_product_df
merged_orders_sales_customers_sellers_installment = pd.merge(
    left=merged_orders_sales_customers_sellers_products,
    right=installment_product_df,
    how='inner',
    on='order_id'
)

# Menggabungkan hasil di atas dengan order_reviews_df
all_df = pd.merge(
    left=merged_orders_sales_customers_sellers_installment,
    right=order_reviews_df,
    how='inner',
    on='order_id'
)
```

### Mengubah frekuensi data untuk memperoleh informasi terkait jumlah order dan total revenue yang diperoleh setiap bulannya

```python
# Menghitung jumlah order dan total revenue per bulan
monthly_orders_df = final_df.resample(rule='M', on='order_approved_at').agg({
    "order_id": "nunique",
    "price": "sum"
})

# Mengubah index menjadi format YYYY-MM
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()

# Mengubah nama kolom untuk hasil akhir
monthly_orders_df.rename(columns={
    "order_approved_at": "order_date",
    "order_id": "order_count",
    "price": "revenue"
}, inplace=True)

print(monthly_orders_df)
```

### Memvisualkan informasi terkait jumlah order dan total revenue yang diperoleh tiap bulan.

## **Kapan terakhir kali pelanggan melakukan transaksi?**

```python
# Mengelompokkan data berdasarkan customer_id dan menghitung nilai RFM
rfm_df = final_df.groupby(by="customer_id", as_index=False).agg({
    "order_approved_at": "max",  # Mengambil tanggal order terakhir
    "order_id": "nunique",        # Menghitung jumlah order unik
    "price": "sum"                # Menghitung total revenue
})

rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]

# Mengambil tanggal terbaru sebagai referensi untuk menghitung recency
recent_date = final_df["order_approved_at"].max()

# Menghitung nilai recency dalam hari
rfm_df["recency"] = (recent_date - rfm_df["max_order_timestamp"]).dt.days

# Menghapus kolom max_order_timestamp karena sudah tidak diperlukan
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
print(rfm_df.sort_values(by="recency", ascending=True).head())
```

Jika hari ini adalah Hari ke-0, maka transaksi terakhir pelanggan tersebut dapat dihitung dengan mengurangi nilai pada kolom "recency" dari hari ini. Berdasarkan data di atas, pelanggan dengan customer_id 28336 memiliki recency 0.0, artinya pelanggan ini melakukan transaksi terakhir pada hari ini. Pelanggan lainnya (customer_id 74186, 85866, 51580, dan 41244) memiliki recency 5.0, yang berarti mereka terakhir kali melakukan transaksi 5 hari yang lalu. Pelanggan yang terakhir kali melakukan transaksi adalah customer_id 28336 yang transaksinya terjadi pada hari ini (recency = 0).
